# Applying Functions

```python
def rating_func(x):
    if x>=8:
        return "Good"
    else:
        return "Bad"
```

```python
movies_df["rating_category"] = movies_df["rating"].apply(rating_func)
movies_df.head(8)

#Or

movies_df["rating_category"] = movies_df["rating"].apply(lambda x : "Good" if x >= 8 else "Bad")
movies_df.head(5)
```
